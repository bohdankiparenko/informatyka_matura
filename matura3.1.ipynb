```python
FN = "dane.txt" # FN skrót od file name

def numbers_from_string(string):
    numbers = []
    substring = ''

    for chr in string:
        if 48 <= ord(chr) < 58:
            substring += chr
        else:
            if len(substring) > 0:
                numbers.append(substring)
                substring = ''
    if len(substring) > 0:
        numbers.append(substring)

    return numbers

with open(FN) as f:
    line = f.readlines()
    lines = line[0].strip().split(';')

lnumbers = []

for line in lines:
    lnumbers.append(numbers_from_string(line))


counter = 0

for numbers in lnumbers:
    if len(numbers) == 0:
        continue
    else:
        if len(numbers) > 1:
            for number in numbers:
                if len(number) < 2:
                    continue
                if number[:2] == '50':
                    counter += 1
        else:
            if numbers[:2] == '50':
                counter += 1
counter

```
