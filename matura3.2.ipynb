```python
FN = "dane.txt" # FN skrót od file name

def numbers_from_string(string):
    numbers = []
    substring = ''

    for chr in string:
        if 48 <= ord(chr) < 58:
            substring += chr
        else:
            if len(substring) > 0:
                numbers.append(substring)
                substring = ''
    if len(substring) > 0:
        numbers.append(substring)

    return numbers

with open(FN) as f:
    line = f.readlines()
    lines = line[0].strip().split(';')

lnumbers = [numbers_from_string(line) for line in lines]

numbers = [str(number) for number in range(10)]

apper_counter = {

}

for number in numbers:

    counter = 0

    for lnumber in lnumbers:
        if len(lnumbers) == 0:
            continue
        else:
            for strnumber in lnumber:
                counter += strnumber.count(number)

    apper_counter[number] = counter

apper_counter
```
