```python
FN = "dane.txt" # FN skrót od file name

def numbers_from_string(string):
    numbers = []
    substring = ''

    for chr in string:
        if 48 <= ord(chr) < 58:
            substring += chr
        else:
            if len(substring) > 0:
                numbers.append(substring)
                substring = ''
    if len(substring) > 0:
        numbers.append(substring)

    return numbers

with open(FN) as f:
    line = f.readlines()
    lines = line[0].strip().split(';')

lnumbers = [numbers_from_string(line) for line in lines]

telnumbers = []

for lnumber in lnumbers:
    if len(lnumber) > 0:
        for number in lnumber:
            if len(number) == 9:
                telnumbers.append(number)

numbersfromtelnumbers = {

}

for telnumber in telnumbers:
    numbers = set()
    for chr in telnumber:
        numbers.add(chr)
    numbersfromtelnumbers[telnumber] = numbers

numbersfromtelnumbers
```
